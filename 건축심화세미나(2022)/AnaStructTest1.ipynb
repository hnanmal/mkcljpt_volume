```python
from anastruct import SystemElements
```

```python
ss = SystemElements()
```

```python
ss.add_element(location=[[0, 0], [0, 7]])
ss.add_element(location=[[0, 7], [9, 7]])
ss.add_element(location=[[9, 7], [9, 0]])
```

```python
ss.add_support_hinged(node_id=1)
ss.add_support_hinged(node_id=4)
```

```python
ss.q_load(element_id=2, q=-41.4)
ss.solve()
```
