```python
#################################Module For Functional Programing#############################################
curry = lambda f: lambda a,*args: f(a, *args) if (len(args)) else lambda *args: f(a, *args)

filter = curry(filter)
map = curry(map)

def _take(length, iter):
    res = []
    for a in iter:
        res.append(a)
        if len(res) == length:
            return res
        
take = curry(_take)
reduce = curry(reduce)

go = lambda *args: reduce(lambda a,f: f(a), args) ## 함수도 축약 가능 ##
##############################################################################################################
```

```python
def importCSV(_path):
    with open(_path,'r') as f:
        reader  = csv.reader(f) # csv의 행별로 읽어옴
        res = [x for x in reader]
        
    return res
```

```python
def exportCSV(_filename, _datas):
    f = open(_filename, 'w', newline='') # 자동줄바꿈 방지 header 이후 첫행 빈행 생성 방지
    csv_writer = csv.writer(f)
    
    for x in _datas:
        csv_writer.writerow(x)
    f.close()
```

```python
def make_messageDict(oneRow, keyInform):
    keys = list(map(lambda x: x[1], keyInform))
    vals = list(map(lambda x: oneRow[x[0]], keyInform))
    
    return dict(zip(keys,vals))
```

```python
make_messageDict(allDatas_net[0], keyInform)
```

```python
allMsgDicts = list(map(lambda x: make_messageDict(x, keyInform), allDatas_net))
allMsgDicts
```

```python
def chk_sender_namePos(dict_):
    chk_tgt_strs = ['매니저','책임','책임매니저','대리','과장']
    for i in chk_tgt_strs:
        if i in dict_['receiver_name_position']:
            str_origin = dict_['receiver_name_position']
            position_str_idx = str_origin.find(i)
            name = str_origin[:position_str_idx].strip()
            position = str_origin[position_str_idx:].strip()
            dict_['receiver_name'] = name
            dict_['receiver_position'] = position
        else: pass
    return dict_
```

```python
allMsgDicts_ = (lambda col: go\
(
    col,
    map(chk_sender_namePos),
    list
))(allMsgDicts)

allMsgDicts_
```
