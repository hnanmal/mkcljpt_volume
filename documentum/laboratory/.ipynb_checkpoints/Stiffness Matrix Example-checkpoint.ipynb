```python
##########################   조건설정해서 build test


nodenumber = [10,20,30,40,50,60]

boundarynode = [60, 10] #index 아닌 node 번호임
boundarynode.sort()  # 추후 결과값  산정을 위해 sort필요함

elements = [[10,20],[20,40],[20,30],[10,30],[30,40],[40,50],[50,60],[40,60]] # always 2 node 
k_member = [500,400,600,200,800,300,700,500] # 각 element 별 stiffness

# node별 결합되는 element를 list index로 반환
node_connection = []
for i in range(0,len(nodenumber)):
    inode_element = []
    for j in range(0,len(elements)):
        if elements[j][0] == nodenumber[i] or elements[j][1] == nodenumber[i]:
            inode_element.append(j)

    node_connection.append(inode_element)

# print(node_connection)



# 전체 구조물 강성 매트릭스 
Kmatrix = np.zeros((len(nodenumber),len(nodenumber)))
for i in range(0,len(nodenumber)):
    inode_element = node_connection[i]
    for j in range(0,len(nodenumber)): 
        if i == j:
            for k in range(0,len(inode_element)): 
                Kmatrix[i][j] = Kmatrix[i][j] + k_member[inode_element[k]]
        else:
            for ie in range(0,len(elements)): 
                if elements[ie] == [nodenumber[i],nodenumber[j]] or elements[ie] == [nodenumber[j],nodenumber[i]]: 
                    Kmatrix[i][j] = Kmatrix[i][j] - k_member[ie]



print('Kmatrix')
print(Kmatrix) # 앞의 예제로 넣은 k값과 비교시 이상 없음 확인 완료 

# boundary node 와 free node 의 index값을 분리
index_boundarynode = []
index_freenode = list(range(0,len(nodenumber)))
for i in range(0,len(boundarynode)):
    index_boundarynode.append(nodenumber.index(boundarynode[i]))
    index_freenode.remove(index_boundarynode[i])


print(index_boundarynode, index_freenode)



# boundary condition 반영 -> bouday의 deflection 은 0으로 들어가므로,
Kmatrix_Structure = []
for i in range(0,len(nodenumber)-len(boundarynode)):
    Kmatrix_Structure.append([])

for i in range(0,len(index_freenode)):
    for j in range(0,len(index_freenode)):
        Kmatrix_Structure[i].append(Kmatrix[index_freenode[i]][index_freenode[j]])

Kmatrix_Structure = np.array(Kmatrix_Structure)


print("Kmatrix_Structure")
print(Kmatrix_Structure)

# 변위 계산 test 를 위한 force matrix로 추후 node 위치별 하중을 받을 수 있는 함수 / class로 구현 필요 
Force_matrix = np.array([800,700,1200,500])

displacement_free = np.linalg.inv(Kmatrix_Structure) @ Force_matrix

# print(displacement_free)

total_disp = np.zeros((len(nodenumber),1))
for i in range(0,len(index_freenode)):
    total_disp[index_freenode[i]] = displacement_free[i]

print('total_disp')
print(total_disp)



total_force = Kmatrix @ total_disp

print('total_force')
print(total_force)

print("Reaction\n", "R",boundarynode[0],total_force[index_boundarynode[0]])
print(" R",boundarynode[1],total_force[index_boundarynode[1]])
```
