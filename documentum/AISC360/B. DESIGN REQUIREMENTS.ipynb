### 허용응력 설계법 (ASD): Allowable Stress Design
구조물을 사용하는데 적용되는 하중과 재료적인 불확실성 등 여러가지 변수를 고려하여 안전율을 적용시켜 설계하는 방법  
설계하중(사용하중)에 의한 실제 응력이 허용응력을 초과하지 않도록 설계하는 것이 핵심  


우리가 허용응력을 산정하는 방법은 재료에 가해지는 응력을 구하여 안전율로 나누어 주는 방법이다.  
구조물에 따라 다르지만 일반적으로 안전율은 2~3을 둔다.  
이는 하중이나 재료적인 면에서의 불확실성을 고려하여  
**'계산상 나오는 응력의 1/3배만 받을수 있다고 보자'** 라는 것이다.

허용응력설계는 부재의 탄성해석에 근거를 두고 있다.  
부재의  탄성해석은 응력이  변형도에  선형비례하는 것,  
즉 구조체가 후크의 법칙에 따라는 것과  하중을 제거했을 때  
잔류변형의 생김이 없이 변형이 원점으로 되돌아오는 것을 전제로 하고 있다.

### 소요강도

부재와 연결부 설계 시에 하중계수를 고려한 총 설계하중 또는 소요강도는 식 (1.5.3-1)과 같다.  

  

구조물 가설위치의 여러 조건과 구조에 따라 적절한 하중 및 하중의 조합을 선정해야 하며  
모든 하중을 모두 고려할 필요는 없다. 하중조합에는 구조물별 설계기준에서 정의된  
여러 한 계상태별 하중조합을 고려한다.  
다만 도로교의 경우 지진 및 선박충돌에 대한 하중은 별도로 고려한다.  
하중계수들은 최대 하중조합 효과가 계산되도록 선정하며, 각 하중조합에서 정과 부의 극한상태를 모두 검토한다.  
한 하중이 다른 하중의 효과를 감소시키는 하중조합에서는 그러한 하중에 최소의 하중계수를 적용한다.
