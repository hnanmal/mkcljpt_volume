```python
## 엑셀 헤더 열 찾기

def findNotNone(row):
    return list(filter(lambda x: x != None, row))

def countNotNone(vals):
    return list(map(len, (map(findNotNone, vals))))

countNotNone(vals)
```

```python
## None이 아닌 유효 데이터가 갑자기 많아지는 Row 찾기

def find_header_startRow_idx(rowObjs):
    vals = [[x.value for x in y] for y in rowObjs]
    validData_count = countNotNone(vals)
    
    for i,j in enumerate(validData_count):
        if j - (validData_count[i-1]) >= 5:
            return i

startRow_idx = find_header_startRow_idx(rowObjs)
startRow_idx
```

```python
def make_sequential_group(nums):
    res = []
    tmp = []
    enum_list = list(enumerate(nums))
    start_i,start_e = enum_list[0]
    for i,e in enum_list:
        if (e - start_e) == (i-start_i):
            if i == len(nums)-1:
                tmp.append(e)
                res.append(deepcopy(tmp))
            else:
                tmp.append(e)
                
        elif (e - start_e) != (i-start_i):
            if i == len(nums)-1:
                res.append(deepcopy(tmp))
                tmp.clear()
                tmp.append(e)
                res.append(deepcopy(tmp))
            else:
                res.append(deepcopy(tmp))
                tmp.clear()
                start_i,start_e = i,e
                tmp.append(e)
        else:
            res.append(deepcopy(tmp))

    return res

make_sequential_group([2, 3, 4, 262, 263, 264, 265, 266, 267])
```

```python
def find_MergedArea(colObj):
    all_idx_merged =  list(filter(lambda x: x > startRow_idx,
                             (map(lambda x: x[0],
                                  (filter(lambda x: "Merged" in str(x),
                                          enumerate(colObj)))))))
    all_idx_merged.insert(0, startRow_idx)
    sequential_group = make_sequential_group(all_idx_merged)

    return sequential_group

find_MergedArea(colObjs[7])
```

```python
def find_header_endRow_idx(colObjs, startRow_idx):
    MergedAreas = list(map(find_MergedArea, colObjs))
    firstElems_MaybeHeader = map(lambda x: x[0], MergedAreas)
    last_idxs = map(lambda x: x[-1], firstElems_MaybeHeader)
    
    return max(last_idxs)

out = find_header_endRow_idx(colObjs, startRow_idx)
out
```

```python
def find_header_objs(rowObjs, colObjs):
    startRow_idx = find_header_startRow_idx(rowObjs)
    endRow_idx = find_header_endRow_idx(colObjs, startRow_idx)
    range_ = range(startRow_idx, endRow_idx+1)
    find_header = lambda rowObj: map(lambda x: x.value, rowObj)
    all_headers = map(lambda x: list(find_header(rowObjs[x])), range_)
    zip_all_headers = zip(*all_headers)
    
    return zip_all_headers


list(find_header_objs(rowObjs, colObjs))
```

```python
headers = list(find_header_objs(rowObjs, colObjs))
list(enumerate(headers))
```

```python
## group 짓기
def find_GroupIndex_NotNone(headers,lv):
    res = []
    for idx,(h1,h2,h3) in enumerate(headers):
        if (h1,h2,h3)[lv] != None:
            res.append(idx)
    return res
find_GroupIndex_NotNone(headers, 0)
```

```python
idxNotNone_ = find_GroupIndex_NotNone(headers, 0)
idxNotNone = deepcopy(idxNotNone_)
idxNotNone.append(idxNotNone_[-1]+1)
idxNotNone
```

```python
def make_headersGroup(headers, lv, target=[], bln_idx="withIdx"):
    headerWithidx = list(enumerate(headers))
    idxNotNone_ = findGroupIndexNotNone(headers, lv)
    idxNotNone = deepcopy(idxNotNone_)
    idxNotNone.append(idxNotNone_[-1]+1)
    group_ranges = zip(range(len(idxNotNone_)), range(1,len(idxNotNone_)+1))
    group_ranges_idx = map(lambda x: idxNotNone[x[0]:x[1]+1], group_ranges)
    
    if bln_idx == "withIdx":
        return list(map(lambda x: (headerWithidx[x[0]:x[1]]), group_ranges_idx))
    elif bln_idx == "noIdx":
        return list(map(lambda x: (target[x[0]:x[1]]), group_ranges_idx))
```

```python
headers_groups = make_headersGroup(headers, 0)
headers_groups
```

```python
def makeDic(colObjs, group):
    h1 = group[0][1][0]
    onlyHeads = list(map(lambda x: x[1], group))
    if len(group) > 1:
        tree_groups = make_headersGroup(onlyHeads, 1, group,"noIdx")
    else:
        tree_groups = [group]
    
    def sameGroupDic(tree_group):
        res = []
        for i in tree_group:
            dic ={}
            idx = i[0]
            h2,h3 = i[1][1],i[1][2]
            if h3 != None:
                dic[h3] = colObjs[idx]
            else:
                dic[h2] = colObjs[idx]
            res.append(dic)
        return res
    
    res = []
    for i in tree_groups:
        h2 = i[0][1][1]
        dic = {}
        v = sameGroupDic(i)
        dic[h2] = v
        res.append(dic)

    return {h1: res}

makeDic(colObjs, headers_groups[5])
```
