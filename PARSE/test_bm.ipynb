## seaborn 연습

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df['time'].unique()
```

```python
sns.barplot(x='time', y='tip', data=df)
```

```python
sns.stripplot(x='time', y='tip', data=df)
```

```python
sns.violinplot(x='time', y='tip', data=df)
```

```python
sns.boxenplot(x='time', y='tip', data=df)
```

```python
sns.boxplot(x='time', y='tip', data=df)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
```

```python
sns.scatterplot(x='total_bill', y='tip', alpha=.2, data=df)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='time', data=df)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='time',
                palette=['purple', 'green'], data=df)
```

```python
sns.regplot(x='total_bill', y='tip', data=df)
```

```python
sns.regplot(x='total_bill', y='tip', lowess=True, data=df,
            line_kws={'color': 'red'}, scatter_kws={'color': 'grey'})
```

```python
sns.distplot(df['tip'])
```

```python
sns.histplot(df['tip'])
sns.kdeplot(df['tip'])
```

```python
sns.kdeplot(df['tip'])
```

```python
sns.distplot(df['tip'], rug=True)
```
